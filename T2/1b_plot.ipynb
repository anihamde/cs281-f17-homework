```python
import numpy as np
import matplotlib.pyplot as plt
import math
import scipy.integrate
import scipy.special
```

```python
def f(d,D):
    return (2**(1-0.5*D))*(d**(D-1.0))*(math.exp(0.5*(-d**2)))/(scipy.special.gamma(0.5*D))
```

```python
Drange = [1,2,3,4,5,10,20,50,100]
dvec = np.linspace(0,15,100)
y = np.zeros((len(Drange),100))
```

```python
for i in range(len(Drange)):
    for j in range(100):
        y[i][j] = f(dvec[j],Drange[i])
```

```python
for k in range(len(Drange)):
    plt.plot(dvec,y[k],label="D = %s"%Drange[k])
    plt.xlabel("Distance")
    plt.ylabel("PDF")
    plt.legend()
plt.savefig('all.png')
plt.show()
```

```python
for k in range(len(Drange)):
    plt.plot(dvec,y[k],label="D = %s"%Drange[k])
    plt.xlabel("Distance")
    plt.ylabel("PDF")
    plt.legend()
    plt.savefig('D=%s.png'%Drange[k])
    plt.show()
```

```python
for k in range(5):
    plt.plot(dvec,y[k],label="D = %s"%Drange[k])
    plt.xlabel("Distance")
    plt.ylabel("PDF")
    plt.legend()
plt.savefig('first5.png')
plt.show()
```

```python
integvals = np.zeros(len(dvec))
for i in range(len(dvec)):
    integvals[i] = scipy.integrate.quad(lambda d: f(d,100),0,dvec[i])[0]
```

```python
plt.plot(dvec,integvals,label='CDF Plot')
plt.xlabel("Distance")
plt.ylabel("CDF")
plt.legend()
plt.savefig('CDF.png')
plt.show()
```
